# ***Importing the libs***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# ***Openning the file and determining the name of the plot***

```python
file = input(r'Which .txt file-path do you want to open? ')
name = input('Which is your spectra name? ')

try:
    with open(file) as file:
        doc = file.readlines() # read all the file as a list of string being each line an element of the list
        display('Plotting the spectra, go to the end of the Jupyter Notebook and see your result! üòÅ')
        
except FileNotFoundError:
    display('File not found, please run the program again and insert a new path! üò•')
```

# ***Treating the data***

```python
axis = list(tuple(line
                      .rstrip(' ') # makes the tuples have only 2 values inside them
                      .replace(' ',',') # replace the separator between the values by a comma
                      .split(',')) # makes a list of tuples at each end of tuple
                      for line in doc)

axis_remove = axis.pop(0) # removes the firts line of the txt which is not relevant for the plot
x,y = tuple(zip(*axis)) # unpacks the list of tuples allowing the conversion to pd.Series to plot the spectra

x_Serie = pd.Series(x,dtype='float64') # converting the tuples, which elements ar strings, to a pandas Series and to float at the same time
y_Serie = pd.Series(y,dtype='float64') # converting the tuples, which elements ar strings, to a pandas Series and to float at the same time
```

# ***Doing the plot***

```python
spectra = sns.lineplot(
    x=x_Serie,
    y=y_Serie,
    size=0.1,
    legend=None,
    color='red')

spectra.set_xlim(4000,350) # reverses the x-axis making the plot look the conventional way 

spectra.set(xlabel='Frequency (1/cm)',
            ylabel='Transmittance (%)',
            title=name)
plt.show()
```
