# ***Importing the libs***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

# ***Openning the file and determining the name of the plot***

```python
file = input(r'Which .txt file-path do you want to open? ')
name = input('Which is your plot name? ')

try:
    with open(file) as file:
        doc = file.readlines()
        display('Plotting the spectra, go to the end of the Jupyter Notebook and see your result! üòÅ')
        
except FileNotFoundError:
    display('File not found, please run the program again and insert a new path üò•')
```

# ***Treating the data***

```python
axis = list(tuple(line
                      .rstrip(' ')
                      .replace(' ',',')
                      .split(',')) 
                      for line in doc)

x,y = tuple(zip(*axis))

x_Serie = pd.Series(x,dtype='float64')
y_Serie = pd.Series(y,dtype='float64')
```

# ***Doing the plot***

```python
spectra = sns.lineplot(
    x=x_Serie,
    y=y_Serie,
    size=0.1,
    legend=None,
    color='red')

spectra.set_xlim(4000,350)

spectra.set(xlabel='Frequency (1/cm)',
            ylabel='Transmittance (%)',
            title=name)
plt.show()
```
