```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import regex as re

#request and openning
file = input(r'Which .txt file-path do you want to open? ')
name = input('Which is your plot name? ')
with open(file) as file:
    doc = file.readlines()

# treating the data
    #del(re.findall('Y'))
    axis = list(tuple(line
                      .rstrip('\n')
                      .rstrip(' ')
                      .replace(' ',',')
                      .split(',')) 
                      for line in doc)

# unpacking the list of tuples and setting them to x and y  
    x,y = tuple(zip(*axis))

# converting x and y to a Serie to plot after
    x_Serie = pd.Series(x,dtype='float64')
    y_Serie = pd.Series(y,dtype='float64')

# plotting the spectra and adjusting the axis to the values of interest in general
spectra = sns.lineplot(
    x=x_Serie,
    y=y_Serie,
    size=0.1,
    legend=None,
    color='green')

spectra.set_xlim(4000,350)

spectra.set(xlabel='Frequency (1/cm)',
            ylabel='Transmitance (%)',
            title=name)
plt.show()
```
