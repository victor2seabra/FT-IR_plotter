# ***Importing the libs***

```python
import pandas as pd
import re
import plotly.express as px
```

# ***Openning the file and determining the name of the plot***

```python
# interaction with the user to get the data and name it to clarify the final result
file = input(r'Which .txt file-path do you want to open? ')
name = input('Which is your spectra name? ')

try:
    with open(file) as file:
        doc = file.readlines() # read all the file as a list of string being each line an element of the list
        print(
            'Plotting the spectra, go to the end of the Jupyter Notebook and see your result! üòÅ'
        )

# if a given path is wrong a message is printed to the user and the program stop executting
except FileNotFoundError:
    print(
        'File not found, please run the program again and insert a new path! üò•'
    )
    exit()
```

# ***Treating the data***

```python
# removing the spaces and turn the .txt into a list of tuples with x and y values
axis = list(tuple
            (line
                      .rstrip(' ') # makes the tuples have only 2 values inside them
                      .replace(' ',',') # replace the separator between the values by a comma
                      .split(',')) # makes a list of tuples at each end of tuple
                      for line in doc # doc is a list and each element is a line on a string format
           )

# removing the first line of the axis variable once it is a string not a data
for i in axis:
    for e in i:
        if re.search('^#.*UNIT',e):
            axis.remove(i)

# unpacks the list of tuples allowing the conversion to pd.Series to plot the spectra
x,y = zip(*axis)

# converting the tuples, which elements are strings, to a pandas Series and to float at the same time
x_Serie = pd.Series(x,
                    dtype='float64'
                   )

# converting the tuples, which elements are strings, to a pandas Series and to float at the same time
y_Serie = pd.Series(y,
                    dtype='float64'
                   )
```

# ***Doing the plot***

```python
image = px.line(
                x=x_Serie,
                y=y_Serie,
                labels={
                    'x': 'Frequency (cm‚Äì1)',
                    'y': 'Transmittance (%)',
                },
                title=name,
               )

image.update_xaxes(autorange='reversed')
image.show()
```
